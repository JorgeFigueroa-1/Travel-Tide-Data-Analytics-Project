10. Percentage of Zero-Seat Flights Associated with Cancellations

This query computes the proportion of zero-seat flights that correspond to cancellations.
Breakdown includes:

- Flights canceled.

- Flights not canceled.

- Records with missing info.
Provides a quantitative validation of whether zero seats = cancellation in practice.

```python
# Percentage of Zero-Seat Flights Associated with Cancellations
sql_query = """
WITH seats_zero AS (
    SELECT
        f.trip_id,
        s.cancellation
    FROM flights f
    LEFT JOIN sessions s ON f.trip_id = s.trip_id
    WHERE f.seats = 0
)
SELECT
    COUNT(*) AS total_seats_zero,
    COUNT(*) FILTER (WHERE cancellation = true) AS cancellations,
    COUNT(*) FILTER (WHERE cancellation = false) AS no_cancellations,
    COUNT(*) FILTER (WHERE cancellation IS NULL) AS no_info,
    ROUND(100.0 * COUNT(*) FILTER (WHERE cancellation = true) / NULLIF(COUNT(*), 0), 2) AS percentage_cancellations
FROM seats_zero
;
"""
df_cancellation_percentage = pd.read_sql(sql_query, engine)

display(df_cancellation_percentage)
```

4. Recomputing Nights Stayed

We recompute the number of nights based on the difference between check_in and check_out.
This ensures consistency between stay duration and recorded nights.

```python
# Recomputing Nights Stayed Based on Check-In and Check-Out Dates
sql_query = """
SELECT
    trip_id,
    hotel_name,
    check_in_time,
    check_out_time,
    GREATEST((check_out_time::date - check_in_time::date), 0) AS nights_calc
FROM hotels
;
"""
df_nights_calc = pd.read_sql(sql_query, engine)

display(df_nights_calc)
```

5. Identifying Zero-Night Hotel Stays

This query flags hotel records where the check-in and check-out occur on the same calendar date.
These are categorized as zero-night stays, which can represent:

- Day-use bookings (if charged).

- Data entry errors or placeholders (if not charged).

```python
# Identifying Zero-Night Hotel Stays
sql_query = """
SELECT
    trip_id,
    hotel_name,
    check_in_time,
    check_out_time,
    rooms,
    hotel_per_room_usd
FROM hotels
WHERE (check_out_time::date - check_in_time::date) = 0
;
"""
df_zero_night_stays = pd.read_sql(sql_query, engine)

display(df_zero_night_stays)
```

6. Classifying Hotel Stays: Valid vs. Zero-Night (Paid/Unpaid)

This classification separates hotel stays into meaningful categories:

- Zero nights  No charge  likely cancellations or placeholder bookings.

- Zero nights  Paid  possible day-use stays.

- Valid stay  normal bookings with at least one night.

```python
# Classifying Hotel Stays: Valid vs. Zero-Night (Paid/Unpaid)
sql_query = """
SELECT
    trip_id,
    hotel_name,
    check_in_time,
    check_out_time,
    GREATEST((check_out_time::date - check_in_time::date), 0) AS nights_calc,
    rooms,
    hotel_per_room_usd,

    CASE
        WHEN (check_out_time::date - check_in_time::date) = 0
             AND hotel_per_room_usd = 0 THEN 'Zero nights - No charge (likely cancellation or placeholder)'
        WHEN (check_out_time::date - check_in_time::date) = 0
             AND hotel_per_room_usd > 0 THEN 'Zero nights - Paid (possible day-use stay)'
        ELSE 'Valid stay'
    END AS stay_category

FROM hotels
ORDER BY nights_calc ASC, hotel_per_room_usd ASC
;
"""
df_stay_category = pd.read_sql(sql_query, engine)

display(df_stay_category)
```

7. Distribution of Hotel Stay Categories with Pricing Insights

This query classifies hotel stays into categories and summarizes them:

- Zero nights  No charge  likely cancellations or placeholders.

- Zero nights  Paid  possible day-use bookings.

- Valid stay  normal multi-night stays.

For each category, it reports:

- Total records (volume of stays).

- Average price per room (pricing consistency check)..

```python
# Distribution of Hotel Stay Categories with Pricing Insights
sql_query = """
SELECT
    stay_category,
    COUNT(*) AS total_records,
    AVG(hotel_per_room_usd) AS avg_price
FROM (
    SELECT
        GREATEST((check_out_time::date - check_in_time::date), 0) AS nights_calc,
        hotel_per_room_usd,
        CASE
            WHEN (check_out_time::date - check_in_time::date) = 0
                 AND hotel_per_room_usd = 0 THEN 'Zero nights - No charge'
            WHEN (check_out_time::date - check_in_time::date) = 0
                 AND hotel_per_room_usd > 0 THEN 'Zero nights - Paid'
            ELSE 'Valid stay'
        END AS stay_category
    FROM hotels
) t
GROUP BY stay_category
;
"""
df_distribution_stay_category = pd.read_sql(sql_query, engine)

display(df_distribution_stay_category)
```

4. Trip ID vs. Booking Validation

We validate whether sessions without a trip_id also have no associated booking activity.

Valid Booking: A trip_id exists and at least one booking (flight or hotel) is made.

Inconsistent Booking: No trip_id but a booking exists (should not happen).

```python
# Combined Trip ID and Booking Consistency Check: all ok
sql_query = """
SELECT
    COUNT(*) FILTER (WHERE trip_id IS NULL AND flight_booked = false AND hotel_booked = false) AS valid_no_booking,
    COUNT(*) FILTER (WHERE trip_id IS NOT NULL AND (flight_booked = true OR hotel_booked = true)) AS valid_booking,
    COUNT(*) FILTER (WHERE trip_id IS NULL AND (flight_booked = true OR hotel_booked = true)) AS inconsistent_booking,
    COUNT(*) FILTER (WHERE trip_id IS NOT NULL AND flight_booked = false AND hotel_booked = false) AS inconsistent_no_booking
FROM sessions
;
"""
df_combined_booking_check = pd.read_sql(sql_query, engine)

display(df_combined_booking_check)
```

# Creation of the session level and user level tables


**Active Users Enriched Table**

Objective:
Build a consolidated table of active users (defined as those with 7 or more sessions since January 4, 2023), enriched with session, flight, and hotel details.

Process:

1. Filter sessions from 2023 onward  only activity after 2023-01-04.

2. Count sessions per user  select only users with more than 7 sessions.

3. Join user attributes  demographics and home location data from users.

4. Attach session details  booking behavior, discounts, cancellations from sessions.

5. Merge trip-level data  add flight (flights) and hotel (hotels) information.

6. Order by session start  to reflect chronological user activity.

Result:
A comprehensive fact table linking users, their sessions, and corresponding travel bookings.
This table becomes the foundation for user-level behavioral analysis, customer segmentation, and churn/loyalty modeling.

**Pipeline Workflow** (Step by Step)

1.  **Filter sessions from 2023 onward**
     Isolate all user sessions starting from January 1, 2023, to define the modern analysis period.

2.  **Count sessions per user (keep users with 7)**
     Calculate the total sessions for each user and filter to retain only active users with 7 or more sessions.

3.  **Join user attributes (demographics & home location)**
     Enrich the filtered sessions by merging them with the user table to add profile and location data.

4.  **Attach session details (discounts, bookings, cancellations)**
     Merge data from bookings and promotions tables to append commercial details like discounts used and booking status.

5.  **Merge trip-level data (flights + hotels)**
     Combine the dataset with flight and hotel tables to attach comprehensive trip information (e.g., destinations, airlines, stays).

6.  **Order by session start (chronological user activity)**
     Sort all records by user and session start time to create a timeline of each user's activity.

7.  **Final Active Users Enriched Table**
     Produce the final table containing enriched, chronological data for all active users, ready for analysis.

# User Segmentation

**Active Users with Corrected Hotel Nights**

This query builds a comprehensive usersession dataset for active users (7 sessions since Jan 4, 2023), integrating demographics, sessions, flights, and hotels. Hotel nights are recalculated from check-in and check-out dates and corrected using validation rules:

- Negative nights  flagged as Invalid stay (data error).

- Zero nights + no charge + cancelled  Cancelled, no cost.

- Zero nights + paid + not cancelled  Day-use stay (valid).

- therwise  Valid stay.

The result provides a reliable foundation for segmentation by ensuring hotel stay quality checks are applied at the user-session level.

**Pipeline Workflow** (Step by Step)

1.  **Filter sessions (2023 onward)**
     Select all user sessions from January 1, 2023, onward to establish the analysis timeframe.

2.  **Count sessions per user (keep users with 7)**
     Calculate the total number of sessions for each user and filter the dataset to only include active users with 7 or more sessions.

3.  **Join user attributes (demographics & home location)**
     Enrich the session data by joining it with the user table to attach demographic details and home location information.

4.  **Attach session details (discounts, bookings, cancellations)**
     Merge data from bookings and cancellations tables to add details like promo codes used, booking status, and cancellation status to each session.

5.  **Merge hotel stays**
     Join the dataset with the hotel stays table to get check-in and check-out dates.
     **Recompute nights:** Calculate `GREATEST(check_out - check_in, 0)` to ensure accurate, non-negative night counts.
    
    
      **Flag invalid/zero-night cases:** Categorize each stay:

    - Negative nights  Invalid stay

    - Zero nights + no charge + cancelled  Cancelled (valid, no cost)

    - Zero nights + paid + not cancelled  Day-use stay (valid)
    
    - Else  Valid stay

6.  **Merge flight trip details (origin, destination, airline, base fare, etc.)**
     Perform a left join with the flight trips table to attach travel details like origin, destination, airline, and fare to the session record.

7.  **Chronological ordering by session start**
     Sort the entire enriched dataset by user and session start time to create a chronological timeline of activity.

8.  **Final UserSessionHotel corrected dataset**
     Produce the final output: a comprehensive, session-level dataset with validated hotel nights, user details, and flight information, ready for analysis.


```python
sql_query = """
-- First User Base Table (from Jan 4, 2023 onwards, active users > 7 sessions)
WITH sessions_2023 AS (
    SELECT *
    FROM sessions
    WHERE session_start >= '2023-01-04'
),

user_session_counts AS (
    SELECT
        user_id,
        COUNT(*) AS total_sessions
    FROM sessions_2023
    GROUP BY user_id
    HAVING COUNT(*) > 7
)

SELECT
		-- User info
    u.user_id,
    usc.total_sessions,  -- total sessions since Jan 4, 2023
    u.birthdate,
    u.gender,
    u.married,
    u.has_children,
    u.home_country,
    u.home_city,
    u.home_airport,
    u.home_airport_lat,
    u.home_airport_lon,
    u.sign_up_date,

    -- Session info
    s.session_id,
    s.session_start,
    s.session_end,
    s.flight_discount,
    s.hotel_discount,
    s.flight_discount_amount,
    s.hotel_discount_amount,
    s.flight_booked,
    s.hotel_booked,
    s.page_clicks,
    s.cancellation,

    -- Hotel summary (corrected nights + validation flag)
    h.hotel_name,
    GREATEST((h.check_out_time::date - h.check_in_time::date), 0) AS nights,
    h.rooms,
    h.check_in_time,
    h.check_out_time,
    h.hotel_per_room_usd,
    CASE
        WHEN (h.check_out_time::date - h.check_in_time::date) < 0
             THEN 'Invalid stay (negative nights  data error)'
        WHEN (h.check_out_time::date - h.check_in_time::date) = 0
             AND h.hotel_per_room_usd = 0
             AND s.cancellation = true THEN 'Zero nights - Cancelled (no charge)'
        WHEN (h.check_out_time::date - h.check_in_time::date) = 0
             AND h.hotel_per_room_usd > 0
             AND s.cancellation = false THEN 'Zero nights - Paid (day-use stay)'
        ELSE 'Valid stay'
    END AS stay_validation,

    -- Flight summary
    f.trip_id,
    f.origin_airport,
    f.destination,
    f.destination_airport,
    f.seats,
    f.return_flight_booked,
    f.departure_time,
    f.return_time,
    f.checked_bags,
    f.trip_airline,
    f.destination_airport_lat,
    f.destination_airport_lon,
    f.base_fare_usd

FROM sessions_2023 s
JOIN user_session_counts usc ON s.user_id = usc.user_id
JOIN users u ON s.user_id = u.user_id
LEFT JOIN flights f ON s.trip_id = f.trip_id
LEFT JOIN hotels h ON s.trip_id = h.trip_id
ORDER BY s.session_start ASC
LIMIT 1000
;
"""

df_hotels_invalid_nights_validated = pd.read_sql(sql_query, engine)
display(df_hotels_invalid_nights_validated.head())
```

**Comprehensive User & Session Summary Table**

This query builds a comprehensive analytical table that combines:

1. Session-level data:

 All user sessions since Jan 4, 2023, enriched with booking/cancellation flags.

 Includes flight and hotel attributes (dates, spend, discounts, locations).

2. Trip-level data:

Groups sessions into trips, validating cancellations and completed bookings.

Provides metrics like confirmed trips, canceled trips, and completed trips per user.

3. User-level data:

Aggregates engagement and financial metrics:

Total sessions with bookings.

Total session time.

Spend on flights & hotels.

Discounts applied.

Sessions without any booking.

4. Hotel-level data:

Summarizes booking metrics per hotel and city (total bookings, room-nights).

5. Final Output:

Produces a single enriched dataset that combines session, user, trip, and hotel-city metrics.

Ready for segmentation, behavioral analysis, churn modeling, and dashboarding (Tableau/BI tools).

**Pipeline Workflow** (Step by Step)

Heres how the query flows logically:

1. Filter sessions  select all sessions from Jan 4, 2023 onward.

2. Active users  keep only users with more than 7 sessions.

3. Session base  enrich sessions with user, flight, and hotel details.

4. Enrich with booking flags  mark sessions as booked/canceled/not booked.

5. Trip-level aggregation  collapse multiple sessions into single trips.

6. Trip rollup  calculate completed, canceled, and confirmed trips per user.

7. User-level session metrics  total time, spend, discounts, booking counts.

8. Sessions without bookings  flag users who never booked.

9. Hotel-city metrics  aggregate room nights and bookings by hotel and city.

10. Final summary  combine session, user, trip, and hotel-level data.

11. Output table  a single dataset containing all metrics for analysis.

```python
sql_query = """
-- 1. Sessions from 2023
WITH sessions_2023 AS (
    SELECT *
    FROM sessions
    WHERE session_start >= '2023-01-04'
),

-- 2. Users with more than 7 sessions in 2023
user_session_counts AS (
    SELECT
        user_id,
        COUNT(*) AS total_sessions
    FROM sessions_2023
    GROUP BY user_id
    HAVING COUNT(*) > 7
),

-- 3. Base session data (join users, flights, hotels)
session_base AS (
    SELECT
        -- User info
        u.user_id,
        usc.total_sessions,
        u.birthdate,
        u.gender,
        u.married,
        u.has_children,
        u.home_country,
        u.home_city,
        u.home_airport,
        u.home_airport_lat,
        u.home_airport_lon,
        u.sign_up_date,

        -- Session info
        s.session_id,
        s.session_start,
        s.session_end,
        s.flight_discount,
        s.hotel_discount,
        s.flight_discount_amount,
        s.hotel_discount_amount,
        s.flight_booked,
        s.hotel_booked,
        s.page_clicks,
        s.cancellation,

        -- Hotel info (split name/city, nights calc)
        SPLIT_PART(h.hotel_name, ' - ', 1) AS hotel_only_name,
        SPLIT_PART(h.hotel_name, ' - ', 2) AS hotel_city,
        GREATEST((h.check_out_time::date - h.check_in_time::date), 0) AS nights,
        h.rooms,
        h.check_in_time,
        h.check_out_time,
        h.hotel_per_room_usd,

        -- Flight info
        f.trip_id,
        f.origin_airport,
        f.destination,
        f.destination_airport,
        f.seats,
        f.return_flight_booked,
        f.departure_time,
        f.return_time,
        f.checked_bags,
        f.trip_airline,
        f.destination_airport_lat,
        f.destination_airport_lon,
        f.base_fare_usd
    FROM sessions_2023 s
    JOIN user_session_counts usc ON s.user_id = usc.user_id
    JOIN users u ON s.user_id = u.user_id
    LEFT JOIN flights f ON s.trip_id = f.trip_id
    LEFT JOIN hotels h ON s.trip_id = h.trip_id
),

-- 4. Enrich sessions with booking/cancellation flags
session_enriched AS (
  SELECT
    sb.*,
    (COALESCE(sb.flight_booked, FALSE) OR COALESCE(sb.hotel_booked, FALSE)) AS has_booking,
    CASE
      WHEN NOT (COALESCE(sb.flight_booked,FALSE) OR COALESCE(sb.hotel_booked,FALSE)) THEN 'null'
      WHEN sb.cancellation = TRUE  THEN 'cancel'
      WHEN sb.cancellation = FALSE THEN 'not cancel'
      ELSE 'null'
    END AS is_canceled
  FROM session_base sb
),

-- 5. Trips with at least one booking
booked_trips AS (
  SELECT DISTINCT user_id, trip_id
  FROM session_enriched
  WHERE trip_id IS NOT NULL
    AND has_booking = TRUE
),

-- 6. Trip-level cancellation status
trip_rollup AS (
  SELECT
    bt.user_id,
    bt.trip_id,
    BOOL_OR(se.cancellation) AS canceled
  FROM booked_trips bt
  JOIN session_enriched se
    ON se.user_id = bt.user_id AND se.trip_id = bt.trip_id
  GROUP BY bt.user_id, bt.trip_id
),

-- 7. User summary (trip-level metrics)
user_summary_trips AS (
  SELECT
    user_id,
    COUNT(*)                                          AS confirmed_trips,
    COUNT(*) FILTER (WHERE canceled = FALSE)          AS completed_trips,
    COUNT(*) FILTER (WHERE canceled = TRUE)           AS canceled_trips
  FROM trip_rollup
  GROUP BY user_id
),

-- 8. Sessions without any real booking
user_sessions_without AS (
  SELECT user_id, COUNT(*) AS sessions_without_bookings
  FROM session_enriched
  WHERE has_booking = FALSE
    AND trip_id IS NULL
  GROUP BY user_id
),

-- 9. User session metrics (time, spend, discounts)
user_sessions_summary AS (
  SELECT
    user_id,
    COUNT(*) FILTER (WHERE has_booking = TRUE) AS total_user_sessions_with_booking,
    SUM(EXTRACT(EPOCH FROM (session_end - session_start))) AS total_session_time,

    -- Spend metrics (keep NULL if no booking at all)
    CASE
      WHEN SUM(CASE WHEN flight_booked = TRUE THEN 1 ELSE 0 END) = 0 THEN NULL
      ELSE SUM(CASE WHEN flight_booked = TRUE THEN base_fare_usd END)
    END AS total_flight_spend,

    CASE
      WHEN SUM(CASE WHEN hotel_booked = TRUE THEN 1 ELSE 0 END) = 0 THEN NULL
      ELSE SUM(CASE WHEN hotel_booked = TRUE THEN hotel_per_room_usd * rooms * nights END)
    END AS total_hotel_spend,

    -- Discounts
    SUM(COALESCE(flight_discount_amount,0)) AS total_flight_discount,
    SUM(COALESCE(hotel_discount_amount,0)) AS total_hotel_discount
  FROM session_enriched
  GROUP BY user_id
),

-- 10. Metrics per hotel/city
hotel_city_metrics AS (
    SELECT
        hotel_city,
        hotel_only_name,
        COUNT(*) FILTER (WHERE hotel_booked = TRUE) AS total_hotel_bookings,
        SUM(rooms * nights) AS total_room_nights
    FROM session_base
    WHERE hotel_city IS NOT NULL
    GROUP BY hotel_city, hotel_only_name
),

-- 11. Final user-level summary
user_summary_final AS (
  SELECT
    u.user_id,
    COALESCE(ut.confirmed_trips, 0)                   AS confirmed_trips,
    COALESCE(ut.completed_trips, 0)                   AS completed_trips,
    COALESCE(ut.canceled_trips, 0)                    AS canceled_trips,
    COALESCE(usw.sessions_without_bookings,0)         AS sessions_without_bookings,
    COALESCE(uss.total_user_sessions_with_booking, 0) AS total_user_sessions_with_booking,
    COALESCE(uss.total_session_time, 0)               AS total_session_time,
    uss.total_flight_spend,
    uss.total_hotel_spend,
    COALESCE(uss.total_flight_discount, 0)            AS total_flight_discount,
    COALESCE(uss.total_hotel_discount, 0)             AS total_hotel_discount
  FROM (SELECT DISTINCT user_id FROM session_enriched) u
  LEFT JOIN user_summary_trips ut   USING (user_id)
  LEFT JOIN user_sessions_without usw USING (user_id)
  LEFT JOIN user_sessions_summary uss USING (user_id)
)

-- 12. Final SELECT (session + user + hotel metrics)
SELECT
    se.*,

    -- Session duration in seconds
    EXTRACT(EPOCH FROM (se.session_end - se.session_start)) AS session_time_seconds,

    -- Flag if session ended with a booking
    CASE
        WHEN se.flight_booked = TRUE OR se.hotel_booked = TRUE
        THEN 1 ELSE 0
    END AS session_with_booking,

    -- Booking lead time (days before trip start)
    CASE
        WHEN se.flight_booked = TRUE AND se.departure_time IS NOT NULL
            THEN (DATE(se.departure_time) - DATE(se.session_start))
        WHEN se.hotel_booked = TRUE AND se.check_in_time IS NOT NULL
            THEN (DATE(se.check_in_time) - DATE(se.session_start))
        ELSE NULL
    END AS booking_lead_days,

    -- User-level metrics
    usf.confirmed_trips,
    usf.completed_trips,
    usf.canceled_trips,
    usf.sessions_without_bookings,
    usf.total_user_sessions_with_booking,
    usf.total_session_time,
    usf.total_flight_spend,
    usf.total_hotel_spend,
    usf.total_flight_discount,
    usf.total_hotel_discount,

    -- Hotel-level metrics
    hcm.total_hotel_bookings,
    hcm.total_room_nights
FROM session_enriched se
LEFT JOIN user_summary_final usf
       ON usf.user_id = se.user_id
LEFT JOIN hotel_city_metrics hcm
       ON se.hotel_city = hcm.hotel_city
      AND se.hotel_only_name = hcm.hotel_only_name
ORDER BY se.session_start
LIMIT 1000;
"""
df_final_user_segmentation = pd.read_sql(sql_query, engine)

display(df_final_user_segmentation.head())
```
